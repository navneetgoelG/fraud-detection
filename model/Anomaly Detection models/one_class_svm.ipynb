```python
# group the class label
normal = df[df.Class == 0]
fraud = df[df.Class == 1]

print(normal.shape)
print(fraud.shape)
```

```python
contamination = len(df[df['Class']==1])/len(df).__float__()
```

```python
X = df.drop(columns=['Class','Time'])
y = df['Class']
```

```python
y = y.astype(np.int8)
y[y==1] = -1
y[y==0] = 1
```

```python
skf = StratifiedKFold(n_splits=4, random_state=1, shuffle=True)
folds = list(skf.split(X, y))
```

```python
oneSVM_model = OneClassSVM()
```

```python
f1sc = make_scorer(f1_score, pos_label=-1, average='binary')
ps = {
        "gamma":np.linspace(0.00005,0.005, 3),
        "nu":[contamination]
    }
search = GridSearchCV(estimator=oneSVM_model, param_grid=ps, scoring=f1sc, cv=folds)
```

```python
search.fit(X, y)
```

```python
df = pd.DataFrame(search.cv_results_)
df
```
